```python
import pandas as pd
import numpy as np
df_US1 = pd.read_csv("../Raw Data/SPX1.csv")
df_US2 = pd.read_csv("../Raw Data/SPX2.csv")
df_US3 = pd.read_csv("../Raw Data/SPX3.csv")
df_US4 = pd.read_csv("../Raw Data/SPX4.csv")
print("df1_shape : " , df_US1.shape)
print("df1_type : " , type(df_US1))
df_US1.info()

print("df2_shape : " , df_US2.shape)
print("df2_type : " , type(df_US2))
df_US2.info()

print("df3_shape : " , df_US3.shape)
print("df3_type : " , type(df_US3))
df_US3.info()

print("df4_shape : " , df_US4.shape)
print("df4_type : " , type(df_US4))
df_US4.info()
```

```python
df_US1["Date"] = pd.to_datetime(df_US1["Date"], errors='coerce')
df_US2["Date"] = pd.to_datetime(df_US2["Date"], errors='coerce')
df_US3["Date"] = pd.to_datetime(df_US3["Date"], errors='coerce')
df_US4["Date"] = pd.to_datetime(df_US4["Date"], errors='coerce')
```

```python
df_US = pd.merge(df_US1, df_US2, on='Date', how='inner')
df_US = pd.merge(df_US, df_US3, on='Date', how='inner')
df_US = pd.merge(df_US, df_US4, on='Date', how='inner')
df_US.dropna(inplace = True)
df_US.info()
```

```python
df_US['SOFR3M'] = ((100 - df_US['SOFR3M'])/100).astype(float)
df_US['VIX'] = (df_US['VIX']/100).astype(float)
df_US['PX_SPREAD'] = (df_US['PX_BID'] - df_US['PX_ASK']).abs()
df_US.info()
df_US.head(50)
```

```python
output_path = '../Clean Data/US_Cleaned.csv'
df_US.to_csv(output_path, index=False)
```
