```python
import pandas as pd
import numpy as np
df_IN1 = pd.read_csv("../Raw Data/NIFTY1.csv")
df_IN2 = pd.read_csv("../Raw Data/NIFTY2.csv")
df_IN3 = pd.read_csv("../Raw Data/NIFTY3.csv")
df_IN4 = pd.read_csv("../Raw Data/NIFTY4.csv")
print("df1_shape : " , df_IN1.shape)
print("df1_type : " , type(df_IN1))
df_IN1.info()

print("df2_shape : " , df_IN2.shape)
print("df2_type : " , type(df_IN2))
df_IN2.info()

print("df3_shape : " , df_IN3.shape)
print("df3_type : " , type(df_IN3))
df_IN3.info()

print("df4_shape : " , df_IN4.shape)
print("df4_type : " , type(df_IN4))
df_IN4.info()
```

```python
df_IN1["Date"] = pd.to_datetime(df_IN1["Date"], errors='coerce')
df_IN2["Date"] = pd.to_datetime(df_IN2["Date"], errors='coerce')
df_IN3["Date"] = pd.to_datetime(df_IN3["Date"], errors='coerce')
df_IN4["Date"] = pd.to_datetime(df_IN4["Date"], errors='coerce')
```

```python
df_IN = pd.merge(df_IN1, df_IN2, on='Date', how='inner')
df_IN = pd.merge(df_IN, df_IN3, on='Date', how='inner')
df_IN = pd.merge(df_IN, df_IN4, on='Date', how='inner')
df_IN.dropna(inplace = True)
df_IN.info()
```

```python
df_IN['IN003M'] = ((df_IN['IN003M'])/100).astype(float)
df_IN['NIFVIX'] = (df_IN['NIFVIX']/100).astype(float)
df_IN['PX_SPREAD'] = (df_IN['PX_BID'] - df_IN['PX_ASK']).abs()
df_IN.info()
df_IN.head(50)
```

```python
output_path = '../Clean Data/IN_Cleaned.csv'
df_IN.to_csv(output_path, index=False)
```
