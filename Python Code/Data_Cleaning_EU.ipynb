```python
import pandas as pd
import numpy as np
df_EU1 = pd.read_csv("../Raw Data/STOXX1.csv")
df_EU2 = pd.read_csv("../Raw Data/STOXX2.csv")
df_EU3 = pd.read_csv("../Raw Data/STOXX3.csv")
df_EU4 = pd.read_csv("../Raw Data/STOXX4.csv")
print("df1_shape : " , df_EU1.shape)
print("df1_type : " , type(df_EU1))
df_EU1.info()

print("df2_shape : " , df_EU2.shape)
print("df2_type : " , type(df_EU2))
df_EU2.info()

print("df3_shape : " , df_EU3.shape)
print("df3_type : " , type(df_EU3))
df_EU3.info()

print("df4_shape : " , df_EU4.shape)
print("df4_type : " , type(df_EU4))
df_EU4.info()
```

```python
df_EU1["Date"] = pd.to_datetime(df_EU1["Date"], errors='coerce')
df_EU2["Date"] = pd.to_datetime(df_EU2["Date"], errors='coerce')
df_EU3["Date"] = pd.to_datetime(df_EU3["Date"], errors='coerce')
df_EU4["Date"] = pd.to_datetime(df_EU4["Date"], errors='coerce')
```

```python
df_EU = pd.merge(df_EU1, df_EU2, on='Date', how='inner')
df_EU = pd.merge(df_EU, df_EU3, on='Date', how='inner')
df_EU = pd.merge(df_EU, df_EU4, on='Date', how='inner')
df_EU.dropna(inplace = True)
df_EU.info()
```

```python
df_EU['EURIBOR3M'] = ((100 - df_EU['EURIBOR3M'])/100).astype(float)
df_EU['VSTOXX'] = (df_EU['VSTOXX']/100).astype(float)
df_EU['PX_SPREAD'] = (df_EU['PX_BID'] - df_EU['PX_ASK']).abs()
df_EU.info()
df_EU.head(50)
```

```python
output_path = '../Clean Data/EU_Cleaned.csv'
df_EU.to_csv(output_path, index=False)
```
